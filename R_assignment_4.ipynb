### Using GGPlot 2

```r
# Load necessary libraries
library(ggplot2)

# Create a dataset
df <- data.frame(
    category = c("A", "B", "C", "D"),
    value = c(70,50,80,20)  # Proportions for the pie chart
)
# Create the pie chart
ggplot(df, aes(x = "", y = value, fill = category)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar(theta = "y") +
  labs(title = "Pie Chart Example") +
  theme_void()
```

```r
# Load ggplot2 library
library(ggplot2)

# Create the bar chart
ggplot(mtcars, aes(x = factor(cyl))) +
  geom_bar(fill = "steelblue", color = "black") +
  labs(title = "Car Count by Cylinder Type", x = "Number of Cylinders", y = "Count") +
  theme_minimal()
```

```r
# Sample dataset
df <- data.frame(
  category = c("A", "B", "C", "D"),
  value = c(10, 15, 7, 12)
)

# Create bar chart
ggplot(df, aes(x = category, y = value, fill = category)) +
  geom_col() +  # Use geom_col() for pre-summarized data
  labs(title = "Custom Bar Chart", x = "Category", y = "Value") +
  theme_minimal()
```

```r
# Load ggplot2 library
library(ggplot2)

# Create scatter plot
ggplot(mtcars, aes(x = hp, y = mpg)) +
  geom_point(color = "blue", size = 3) +  # Add points with custom color and size
  labs(title = "Scatter Plot of MPG vs Horsepower", x = "Horsepower (hp)", y = "Miles per Gallon (mpg)") +
  theme_minimal()  # Use a clean theme
```
